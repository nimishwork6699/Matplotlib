```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4]) # when we give a single one it will imagine it for the y axis and for the x axis it willl start from 0
plt.show()
```

```python
plt.plot([1,2,3,4],[4,5,6,7]) # here it takes (x,y)
plt.show()
```

```python
x=[1,2,3,4]
y=[90,92,94,96]
plt.plot(x,y)
plt.title("Marks corresponding to no of hours studied")
plt.xlabel("No of hours studied")
plt.ylabel("Marks")
plt.show()
```

```python
x=[1,2,3,4]
y=[90,92,94,96]
plt.plot(x,y)
plt.title("Marks corresponding to no of hours studied")
plt.xlabel("No of hours studied")
plt.ylabel("Marks")
plt.xticks([1,2,4])     # we will only show the no that we want in the pic to be shown
plt.yticks([90,94,100])
plt.show()
```

```python
x=[1,2,3,4]
dsa_marks=[90,92,94,96]
ml_marks=[90,96,97,100]

plt.plot(x,dsa_marks , label="dsa_marks",color="r",linewidth=2,marker="*",linestyle="--")
plt.plot(x,ml_marks, label="ml_marks")

plt.title("Marks corresponding to no of hours studied",
          fontdict={'fontname':'Ariel','fontweight':20,'fontsize':20,'color':'blue'})
plt.xlabel("No of hours studied")
plt.ylabel("Marks")

plt.xticks([1,2,3,4])     # we will only show the no that we want in the pic to be shown
plt.yticks([90,92,94,96,100])

plt.legend(loc="lower right")
plt.savefig("marks.png",dpi=1000,transparent=)
plt.show()
```

```python
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
t = np.arange(0., 5., 0.2)

plt.figure
plt.plot(t, t, 'r--')
plt.show()

plt.figure
plt.plot( t, t**2, 'bs')
plt.show()

plt.figure
plt.plot( t, t**3, 'g^')
plt.show()
```

```python
t = np.arange(0., 5., 0.2)

plt.figure

plt.subplot(1,3,1) #no of rows,no of colums , index of them
plt.plot(t, t, 'r--')
plt.xlabel("T values")
plt.ylabel("Outout values")

plt.title("Linear")

plt.subplot(1,3,2)
plt.plot( t, t**2, 'bs')
plt.xlabel("T values")
plt.ylabel("Outout values")

plt.title("Quadratic")

plt.subplot(1,3,3)
plt.plot( t, t**3, 'g^')
plt.xlabel("T values")
plt.ylabel("Outout values")

plt.title("Cubic")

plt.suptitle("Diff function of t")
plt.tight_layout() # for the padding


plt.show()
```

```python
t = np.arange(0., 5., 0.2)

plt.figure

plt.subplot(3,1,1) #no of rows,no of colums , index of them
plt.plot(t, t, 'r--')

plt.subplot(3,1,2)
plt.plot( t, t**2, 'bs')

plt.subplot(3,1,3)
plt.plot( t, t**3, 'g^')
plt.show()
```

```python
# axes- x axis , y axis, labels,tickers

x=np.linspace(0,10,100) # 0-10 ke bich will create 100 equal points

fig,axis=plt.subplots(2,3) # rows,colums

axis[0,0].plot(x,np.sin(x))
axis[0,0].set_title("sin funct")
axis[0,0].set_xlabel('x')
axis[0,0].set_ylabel('y')

axis[0,1].plot(x,np.cos(x))
axis[0,1].set_title("cos funct")
axis[0,1].set_xlabel('x')
axis[0,1].set_ylabel('y')

# and so on

plt.tight_layout()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Scatter Plot

```python
x=np.random.rand(100)*10
y=np.random.rand(100)*10

plt.scatter(x,y,marker='D',edgecolors='black',linewidths=2,alpha=0.3,s=50,label='random')
# with the help of alpha we can see the overlapping
plt.legend(loc="lower right")
plt.grid()
plt.show()
```

# BAR Plot

```python
courses=['GenAI','DSA','RAG','PIP','GIT']
students=[2700,2400,3100,2100,2500]

plt.bar(courses,students,width=0.5 , align='edge',edgecolor='black',linewidth=2,color='red',hatch='*')
# align is for the text below
plt.show()
```

```python
courses=['GenAI','DSA','RAG','PIP','GIT']
students=[2700,2400,3100,2100,2500]
hatches=['/','|','o','*',"//"]

bars=plt.bar(courses,students,width=0.5 , align='edge',edgecolor='black',linewidth=2,color='red')

for bar,hatch in zip (bars,hatches):
  bar.set_hatch(hatch)
# align is for the text below
plt.show()
```

# Histogram

```python
ages=[21,24,34,56,46,68,66,78,90] # it shows with ages and to the frequency it occurs

sorted_ages=np.sort(ages)
print(sorted_ages)

plt.hist(ages,bins=5,orientation="horizontal",edgecolor="black",color="red")
plt.xlabel("Frequency")
plt.ylabel("Ages")

plt.show()
```

# Pie plot

```python
courses=['GenAI','DSA','RAG','PIP','GIT']
students=[2700,2400,3100,2100,2500]
explodes=[0.2,0,0,0,0]
pie_color=['red','blue','yellow','green','purple']

plt.pie(students,labels=courses,shadow=True,explode=explodes,colors=pie_color,autopct='%.2f%%')
plt.legend(loc="lower right")
plt.show()
```

# 3d scatter plot

```python
ax=plt.axes(projection="3d")

x=np.random.rand(100)
y=np.random.rand(100)
z=np.random.rand(100)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax.scatter(x,y,z)
plt.show()
```

# 3d plots

```python
ax=plt.axes(projection="3d")

x=np.arange(0,50,0.1)
y=np.sin(x)
z=np.cos(x)

ax.plot(x,y,z)
plt.show()
```

```python
ax=plt.axes(projection="3d")

x=np.random.randint(1,10,5)
y=np.random.randint(1,10,5)
z=np.zeros(5)

dx=np.ones(5)
dy=np.ones(5)
dz=np.random.randint(1,10,5)


ax.bar3d(x,y,z,dx,dy,dz)
plt.show()
```

# Plot Surface

```python
ax=plt.axes(projection="3d")

x=np.arange(-5,5,0.5)
y=np.arange(-5,5,0.5)

x,y=np.meshgrid(x,y)
r=np.sqrt(x**2 + y**2)
z=np.sin(r)

ax.plot_surface(x,y,z,cmap="plasma")

plt.show()

```
